## محاسبه شباهت و تفاوت ارای دو سناتور . اولی و دومی

```python
from sklearn.metrics.pairwise import euclidean_distances
euclidean_distances(data.iloc[0,3:], data.iloc[1,3:])
```

## Clustering

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2, random_state=1)
similarities = model.fit_transform(data.iloc[:, 3:])
similarities
```

## استخراج لیبل ها از مدل
## lables_

```python
labels = model.labels_
labels
```

```python
print(pd.crosstab(labels, data["party"]))
```

## بررسی 3 سناتور خارج از کلاستر

## Plot

```python
import matplotlib.pyplot as plt
plt.scatter(x=similarities[:,0], y=similarities[:,1], c=labels, linewidths=0)
plt.show()
```
