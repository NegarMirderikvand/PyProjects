## Clustering

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2, random_state=1)
similarities = model.fit_transform(data.iloc[:, 3:])
similarities
```

## Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ù„ÛŒØ¨Ù„ Ù‡Ø§ Ø§Ø² Ù…Ø¯Ù„
## lables_

```python
labels = model.labels_
labels
```

```python
print(pd.crosstab(labels, data["party"]))
```
