```python
categoricals = df_not_Empty.select_dtypes(include=['object']).columns
for col in categoricals:
    print(col+":", len(train[col].unique()))
```

```python
for col in categoricals:
    train[col] = train[col].astype('category')
```

## ِDummy Vars

```python
dummy_cols = pd.DataFrame()

for col in categoricals:
    col_dummies = pd.get_dummies(train[col])
    train = pd.concat([train, col_dummies], axis=1)
    del train[col]
train
```

## یافت و مدیریت مقادیر از دست رفته

```python
train_novalue = train.isnull().sum()
df_missings = train[train_novalue[(train_novalue>0) & (train_novalue<584)].index]
print(df_missings.isnull().sum())
print(df_missings.dtypes)
```
