## پرکردن مقادیر خالی داده های اسمی و کتگوریکال

## شروع میشوند CONTRIBUTING ستون هایی که با

```python
crashSotuns = [c for c in crash.columns if c.startswith("CONTRIBUTING")]

responsibles = crash[crashSotuns]
```

## تبدیل چند ستون  به یک سری 

```python
responsiblesID = responsibles.stack()
```

## شمارش دلیل هر تصادف

```python
vehicles_counts = responsiblesID.value_counts()
top10_crash = vehicles_counts.head(10)
top10_crash
```
