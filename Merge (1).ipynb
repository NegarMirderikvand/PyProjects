```python
import pandas as pd
```

```python
devices = pd.read_csv("C:/Users/ABA/Desktop/nikamooz/mainCourseSTuuf/dataset/devices.csv", index_col=False)
userUsage = pd.read_csv("C:/Users/ABA/Desktop/nikamooz/mainCourseSTuuf/dataset/userUsage.txt", index_col=False)
userDev = pd.read_csv("C:/Users/ABA/Desktop/nikamooz/mainCourseSTuuf/dataset/user-device.csv", index_col=False)
devices.head()
```

```python
userDev.head()
```

## Left Merge

```python
result = pd.merge(userUsage,
                 userDev[['use_id', 'platform', 'device']],
                 on='use_id', 
                 how='left')
result.head(10)
```

## Right Merge

```python
pd.merge(userUsage,
                 userDev[['use_id', 'platform', 'device']],
                 on='use_id', 
                 how='right')
```

## Full Outer Merge

```python
pd.merge(userUsage,
                 userDev[['use_id', 'platform', 'device']],
                 on='use_id', 
                 how='outer', 
                 indicator=True)
```

## Inner Merge

```python
pd.merge(userUsage,
                 userDev[['use_id', 'platform', 'device']],
                 on='use_id', 
                 how='inner', left_index= False, right_index = False )
```
