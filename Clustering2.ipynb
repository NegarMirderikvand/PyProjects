```python
Basketballdf = pd.read_csv('C:/Users/ABA/Desktop/nikamooz/mainCourseSTuuf/dataset/players_stats.csv')
Basketballdf
```

```python
Forsat_saz = Basketballdf[Basketballdf['Pos'] == 'PG']
Forsat_saz = Forsat_saz[Forsat_saz['TOV'] != 0]
Forsat_saz =  Forsat_saz[Forsat_saz['Games Played']!= 0] 
Forsat_saz= Forsat_saz[Forsat_saz['PTS']!=0]
Selected_ForsatSaz = Forsat_saz[['Name','Games Played','PTS','FG%','FT%']]
Selected_ForsatSaz
```

```python
Selected_ForsatSaz['ppg'] = Selected_ForsatSaz['PTS'] / Selected_ForsatSaz['Games Played']
Selected_ForsatSaz
```

```python
Forsat_saz = Forsat_saz[Forsat_saz['TOV'] != 0]
Selected_ForsatSaz['AST/TOV'] = Forsat_saz['AST']/ Forsat_saz['TOV']
Selected_ForsatSaz.fillna(0)
```

##  PLOT

```python
import matplotlib.pyplot as plt
plt.scatter(Selected_ForsatSaz['ppg'], Selected_ForsatSaz['AST/TOV'], c='y')
plt.title("Forsat Saz")
plt.xlabel('Game Score', fontsize=13)
plt.ylabel('AST/TOV', fontsize=13)
plt.show()
```

```python
def visualize_clusters(df, num_clusters):
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    for n in range(num_clusters):
        clustered_df = df[df['cluster'] == n]
        plt.scatter(clustered_df['ppg'], clustered_df['AST/TOV'], c=colors[n-1])
        plt.xlabel('Points Per Game', fontsize=13)
        plt.ylabel('Assist Turnover Ratio', fontsize=13)
    plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(Selected_ForsatSaz[['ppg', 'AST/TOV']])
Selected_ForsatSaz['cluster'] = kmeans.labels_
visualize_clusters(Selected_ForsatSaz, 5)
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))

visualizer.fit(Selected_ForsatSaz[['ppg', 'AST/TOV']])
visualizer.show()
```
