```python
import pandas as pd
import numpy as np
```

```python
rentals = pd.read_csv("C:/Users/ABA/Desktop/nikamooz/mainCourseSTuuf/dataset/listings.csv")
```

```python
selectedRentals = rentals[['host_listings_count','latitude','longitude','accommodates','bedrooms','bathrooms','beds','price']]
selectedRentals
```

## Cleaning-Preparation

```python
selectedRentals.fillna(0, inplace=True)
selectedRentals
```

```python
first_living_space_value = selectedRentals.iloc[0]['accommodates']
first_living_space_value
```

## محاسبه تفاوت ملک مدنظر با بقیه املاک از نظر ظرفیت نفرات

```python
our_accommodation = 3
selectedRentals['distance'] = selectedRentals['accommodates'].apply(lambda x: np.abs(x - our_accommodation))
print(selectedRentals['distance'].head(10))
```

## از هر عدد فاصله چند تا در محاسبه ما وجود دارد

```python
print(selectedRentals['distance'].value_counts())
```

```python
print(selectedRentals[selectedRentals["distance"] == 0]["accommodates"])
```

## پاکسازی فیلد قیمت
## میانگین قیمت 5 خانه

```python
stripped_commas = selectedRentals['price'].str.replace(',', '')
stripped_dollars = stripped_commas.str.replace('$', '')
selectedRentals['price'] = stripped_dollars.astype('float')
mean_price = selectedRentals.iloc[0:5]['price'].mean()
print(mean_price)
```
