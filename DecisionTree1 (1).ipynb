```python
import pandas as pd
import numpy as np
```

```python
dataSet = pd.read_csv('C:/Users/ABA/Desktop/nikamooz/mainCourseSTuuf/dataset/adults/adult.dat')
```

## Categorical to Numerical

```python
columnss = pd.Categorical(dataSet["workclass"])
dataSet["workclass"] = columnss.codes
for i in ["education", "marital-status", "occupation", "relationship", "race", "sex", "native-country", 'Lable']:
    columnss = pd.Categorical(dataSet[i])
    dataSet[i] = columnss.codes
dataSet
```

## Fit the tree

```python
from sklearn.tree import DecisionTreeClassifier

Needed_columns = ["age", "workclass", "education-num", "marital-status", "occupation", "relationship",
                  "race", "sex", "hours-per-week", "native-country"]

clf = DecisionTreeClassifier(random_state=1)

clf.fit(dataSet[Needed_columns], dataSet["Lable"])
```

## Train-Test Data

```python
import math
np.random.seed(1)

dataSet = dataSet.reindex(np.random.permutation(dataSet.index))
#permutation: permute rows

```

## AUC in Test

```python
from sklearn.metrics import roc_auc_score
train_max_row = math.floor(dataSet.shape[0] * .8)
#floor: = or < 
train = dataSet.iloc[:train_max_row]
test = dataSet.iloc[train_max_row:]
model = DecisionTreeClassifier(random_state=1)
model.fit(train[Needed_columns], train["Lable"])
predictions = model.predict(test[Needed_columns])
AUC = roc_auc_score(test["Lable"], predictions)
print(AUC)
```

## AUC in Train

```python
predictions = model.predict(train[Needed_columns])
print(roc_auc_score(train["Lable"], predictions))
```

## plot Tree

```python
from sklearn import tree
import pydotplus
from IPython.display import Image 
```

```python
dot_data = tree.export_graphviz(model, out_file=None, 
                                feature_names=Needed_columns,  
                                class_names='1,2')

graph = pydotplus.graph_from_dot_data(dot_data)  


Image(graph.create_png())
```

## Limit Overfit

```python
model2 = DecisionTreeClassifier(random_state=1)
model2 = DecisionTreeClassifier(min_samples_split=3, random_state=1)
model2.fit(train[Needed_columns], train["Lable"])
predictions = model2.predict(test[Needed_columns])
test_auc = roc_auc_score(test["Lable"], predictions)
train_predictions = model2.predict(train[Needed_columns])
train_auc = roc_auc_score(train["Lable"], train_predictions)
print(test_auc)
print(train_auc)
```

## Min Samples Split 

```python
model2 = DecisionTreeClassifier(random_state=1)
model2 = DecisionTreeClassifier(min_samples_split=13, random_state=1)
model2.fit(train[Needed_columns], train["Lable"])
predictions = model2.predict(test[Needed_columns])
test_auc = roc_auc_score(test["Lable"], predictions)
train_predictions = model2.predict(train[Needed_columns])
train_auc = roc_auc_score(train["Lable"], train_predictions)
print(test_auc)
print(train_auc)
```

## Depth

```python
model2 = DecisionTreeClassifier(random_state=1)
model2 = DecisionTreeClassifier(min_samples_split=13,max_depth=5, random_state=1)
model2.fit(train[Needed_columns], train["Lable"])
predictions = model2.predict(test[Needed_columns])
test_auc = roc_auc_score(test["Lable"], predictions)
train_predictions = model2.predict(train[Needed_columns])
train_auc = roc_auc_score(train["Lable"], train_predictions)
print(test_auc)
print(train_auc)
```

## Depth 2
## Min Sample 90

```python
model2 = DecisionTreeClassifier(random_state=1)
model2 = DecisionTreeClassifier(min_samples_split=90,max_depth=2, random_state=1)
model2.fit(train[Needed_columns], train["Lable"])
predictions = model2.predict(test[Needed_columns])
test_auc = roc_auc_score(test["Lable"], predictions)
train_predictions = model2.predict(train[Needed_columns])
train_auc = roc_auc_score(train["Lable"], train_predictions)
print(test_auc)
print(train_auc)
```
