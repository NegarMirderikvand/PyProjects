```python
dc_listings = rentals.loc[np.random.permutation(len(rentals))]
removeComma = dc_listings['price'].str.replace(',', '')
removeDollar = removeComma.str.replace('$', '')
dc_listings['price'] = removeDollar.astype('float')
print(dc_listings.info())
```

## حذف ستون زائد

```python
drop_columns = ['room_type', 'city', 'state', 'latitude', 'longitude', 'zipcode',
                'host_response_rate', 'host_acceptance_rate',
                'host_listings_count']
dc_listings = dc_listings.drop(drop_columns, axis=1)
print(dc_listings.isnull().sum())
```

```python
dc_listings = dc_listings.drop(['cleaning_fee', 'security_deposit','license','notes'], axis=1)
dc_listings = dc_listings.dropna(axis=0)
print(dc_listings.isnull().sum())
```

## انتخاب داده مورد نیاز

```python
dc_listings = rentals.loc[np.random.permutation(len(rentals))]
stripped_commas = dc_listings['price'].str.replace(',', '')
stripped_dollars = stripped_commas.str.replace('$', '')
dc_listings['price'] = stripped_dollars.astype('float')
selectedList = dc_listings[['accommodates','bedrooms','bathrooms','beds','minimum_nights','maximum_nights','number_of_reviews']]
selectedList.sort_index().head(10)
```

## حذف مقادیر خالی

```python
selectedList.fillna(0)
selectedList.fillna(0,inplace=True)
selectedList.sort_index().head()
```

## Normalization (Single Column)

```python
first_transform = selectedList['maximum_nights'] - selectedList['maximum_nights'].mean()
normalized_col = first_transform / first_transform.std()
```

## Normalization (DataFrame)

```python
normalized_df = (selectedList - selectedList.mean()) / (selectedList.std())
```

```python
normalized_List = (selectedList - selectedList.mean())/(selectedList.std())
normalized_List['price'] = dc_listings['price']
print(normalized_List.head(3))
```

## فاصله اقلیدسی

```python
from scipy.spatial import distance
List1 = [1, 1]
List2 = [-1, -1]
dist = distance.euclidean(List1, List2)
dist
```

## Multi Criteria KNN

```python
first = normalized_List.iloc[0][['beds', 'bathrooms']]
fifth = normalized_List.iloc[4][['beds', 'bathrooms']]
first_fifth_distance = distance.euclidean(first, fifth)
print(first_fifth_distance)
```

## Constructor and Instantiate 

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
```

## Train and Test Set

```python
# Split full dataset into train and test sets.
train_df = normalized_List.iloc[0:2792]
test_df = normalized_List.iloc[2792:]
```

## Prediction

```python
train_columns = ['beds', 'bathrooms']
knn = KNeighborsRegressor(n_neighbors=5, algorithm='brute', metric='euclidean')
knn.fit(train_df[train_columns], train_df['price'])
predictions = knn.predict(test_df[train_columns])
```

## Evaluation ارزیابی

```python
from sklearn.metrics import mean_squared_error
```

```python
two_features_mse = mean_squared_error(test_df['price'], predictions)
two_features_rmse = two_features_mse ** (1/2)
print(two_features_mse)
print(two_features_rmse)
```
