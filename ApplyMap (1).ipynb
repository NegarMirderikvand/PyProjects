## low/high توابع

```python
def transformer (cell):
    if cell > 1 :
        return 'High'
    else:
        return 'Low'
```

```python
def transformer1 (e,m):
    if e > m :
        return 'High'
    else:
        return 'Low'
```

## Apply

```python
ecoApply =data['Economy'].apply(transformer)
```

## Map 
## به خطا توجه کنید

```python
ecoMap = data['Economy'].map(transformer1, m = data['Economy'].mean())
```

## محاسبه میانگین اقتصاد و
## لیبل گذاری روی بیشتر از میانگین و کمتر از میانگین

```python
ecoMap = data['Economy'].apply(transformer1, m = data['Economy'].mean())
ecoMap.head()
```

## سنجش تساوی دو سریز

```python
equal = ecoApply.equals(ecoMap)
equal
```

## درصد تاثیر بعضی ستون ها
